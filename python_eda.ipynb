```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('netflix.csv')
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
df.info()
```

```python
df.describe()
```

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df['director'].replace(np.nan, 'No Director',inplace=True)
df['cast'].replace(np.nan, 'No Cast',inplace=True)
df['country'].replace(np.nan, 'Not Specify',inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 4))
sns.boxplot(x=df['release_year'])
plt.title('Boxplot for Release Year to Check for Outliers')
plt.show()
```

```python
plt.title('Movies vs Tv Shows')
sns.countplot(x='type', data=df)
```

```python
df_rating = df['rating'].value_counts()
df_rating = pd.DataFrame(df_rating).reset_index()
df_rating.columns = ['rating','Nbr']
plt.title('Most popular ratings in netflix')
sns.barplot(x = 'rating',y = 'Nbr', data=df_rating.head(5));
```

```python
plt.figure(figsize = (12,8))
sns.countplot(x='rating',data = df,hue='type')
```

```python
# Distplot for release_year
plt.figure(figsize=(10, 6))
sns.distplot(df['release_year'], kde=True, bins=30)
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Density')
plt.show()

# Histogram for release_year
plt.figure(figsize=(10, 6))
plt.hist(df['release_year'], bins=30, edgecolor='black')
plt.title('Histogram of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.show()
```

```python
# The distribution of release years is right-skewed, indicating that most of the content on Netflix is relatively new, with a significant amount released in the last decade.
```

```python
sns.countplot(y='listed_in', order=df['listed_in'].value_counts().index[0:10],data=df)
plt.title("Top 10 Genres streamed on Netflix")
```

```python
plt.figure(figsize=(12,6))
df[df["type"]=="Movie"]["listed_in"].value_counts()[:10].plot(kind="barh",color="Red")
plt.title("Top 10 Genres of Movies",size=18)
```

```python
plt.figure(figsize=(12,6))
df[df["type"]=="TV Show"]["listed_in"].value_counts()[:10].plot(kind="barh",color="black")
plt.title("Top 10 Genres of TV Shows",size=18)
```

```python
plt.figure(figsize=(6,6))
y = sns.countplot(y='country',
            data=df, order =df['country'].value_counts().index[:8])

plt.tick_params(labelsize=8)
plt.title("Country With Highest number of shows")
plt.show()
```

```python
# Countplot for top 10 release years
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='release_year', order=df['release_year'].value_counts().iloc[:10].index)
plt.title('Top 10 Most Frequent Release Years')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

```python
tv_shows_df = df[df['type'] == 'TV Show']

# Create a countplot to compare the number of seasons in TV shows
plt.figure(figsize=(18, 10))
sns.countplot(x='duration', data=tv_shows_df, order=tv_shows_df['duration'].value_counts().index)
plt.title("Comparison of the Number of Seasons in TV Shows on Netflix")
plt.xlabel("Number of Seasons")
plt.ylabel("Count of TV Shows")
plt.show()

```

```python
most_titles_produced = df.groupby('country').count().sort_values('title',ascending=False).head(5)

most_titles_produced.reset_index(inplace=True)
most_titles_produced
plt.bar(most_titles_produced.country,most_titles_produced.title);
```

```python
from sklearn import preprocessing
le_features = ['show_id', 'type', 'title', 'director', 'cast', 'country','rating', 'duration', 'listed_in', 'description']
le = preprocessing.LabelEncoder()
for i in le_features:
    df[i] = le.fit_transform(df[i])
# applying label encoding to convert object datatype to numeric datatype
```

```python
correlation_matrix = df.corr()
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True,cmap="mako")
plt.title("Correlations Between Variables",size=15)
plt.show()
```
